### 2. Standardization / Normalization

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
```

```python
std_scalar = StandardScaler()
norm_scalar = Normalizer()

col = df[["Umur", "Berat Akhir", "Jmlh Tahun Pendidikan", "Keuntungan Kapital", "Kerugian Capital", "Jam per Minggu"]]
data_scaler = std_scalar.fit_transform(col)
```

### 3. Preprocessing Data

#### 3.1 Imputing Missing Value

```python
# melihat 5 data teratas
df.head()
```

```python
df["Kelas Pekerja"].value_counts()
```

```python
df["Pendidikan"].value_counts()
```

```python
df["Pekerjaan"].value_counts()
```

```python
#mengubah nilai "?" yang terdapat pada column "Kelas Pekerja" dan "Pekerja"
df["Kelas Pekerja"] = df["Kelas Pekerja"].replace({"?" : np.nan})
df["Pekerjaan"] = df["Pekerjaan"].replace({"?" : np.nan})

# filling nan value with modus
choices_1 = ["Wiraswasta", "Pekerja Bebas Bukan Perusahan", "Pemerintah Lokal"]
df["Kelas Pekerja"] = df["Kelas Pekerja"].fillna(method="ffill")
choices_2 = ["Spesialis", "Perbaikan Kerajinan", "Ekesekutif Managerial", "Pemuka Agama", "Sales"]
df["Pekerjaan"] = df["Pekerjaan"].fillna(method="ffill")
```

```python
# ngecek tipis
df["Kelas Pekerja"].value_counts()
```

```python
#ngecek tipis
df["Pekerjaan"].value_counts()
```

#### 3.2 Encoding Categorial Data

```python
# Mendummies column "Status Perkawinan", "Pekerjaan", "Jenis Kelamin" karena bertipe non-ordinal
df = pd.get_dummies(df, columns=["Kelas Pekerja", "Status Perkawinan", "Pekerjaan", "Jenis Kelamin"])

df.head()
```

```python
# Mereplace column "Pendidikan", dan "Gaji" karena bertipe ordinal
col_1 = {
    "Doktor" : 15, "Master" : 14, "Sekolah Professional" : 13, "Sarjana" : 12,
    "Pendidikan Tinggi" : 11, "D4" : 10, "D3" : 9, "SMA" : 8, "12th" : 7,
    "11th" : 6, "10th" : 5, "9th" : 4, "7th-8th" : 3, "5th-6th" : 2,
    "1st-4th" : 1, "SD" : 0
}

col_2 = {
    ">7jt" : 1,
    "<=7jt" : 0
}

df["Pendidikan"] = df["Pendidikan"].replace(col_1)
df["Gaji"] = df["Gaji"].replace(col_2)

df
```

```python
# mengambil column yang tidak masuk di standardisasi
df_partial = df.drop(["Umur", "Berat Akhir", "Jmlh Tahun Pendidikan", "Keuntungan Kapital",
                      "Kerugian Capital", "Jam per Minggu"], axis=1)

df_partial
```

#### 3.3 Merging DataFrame

```python
# Membuat data frame dari data_scaler
df_scaler = pd.DataFrame(data_scaler, columns=["Umur", "Berat Akhir", "Jmlh Tahun Pendidikan",
                                              "Keuntungan Kapital", "Kerugian Capital", "Jam per Minggu"])
df_scaler["id"] = df["id"]

df_scaler
```

```python
# menggabungkan dataset df_scaler dan df_partial
df_new = pd.merge(df_scaler, df_partial, on="id")
df_new
```

### 4. Modelling

#### 4.1 Membuat Algoritma Machine Learning

```python
# import libary
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

# memisahkan data feature dan data target (label)
X = df_new.drop(["id", "Gaji"], axis=1)
y = df_new["Gaji"]

# membuat objek untuk model selection
model = KNeighborsClassifier()

# membuat model untuk melihat kombinasi terbaik 
param_grid = {
    "n_neighbors" : np.arange(5, 21),
    "weights" : ["uniform", "distance"],
    "p" : [1, 2]
}
gscv = GridSearchCV(model, param_grid=param_grid, scoring="roc_auc", cv=10)
gscv.fit(X, y)
```

```python
# mengecek kombinasi terbaik berdasarkan hyperparameter tuning di atas
gscv.best_params_
```

```python
# mengecek skor dari kombinasi terbaik
gscv.best_score_
```
